##**Facial Detection**

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images_Test/people2.jpg')

face_detector_cnn = dlib.cnn_face_detection_model_v1('/content/drive/MyDrive/Colab Notebooks/CNN(convolutional neural network)_detection/Weights/mmod_human_face_detector.dat')
detections = face_detector_cnn(image, 1)
```

```python
for face in detections:
  print(face.confidence)

len(detections)
```

```python
for face in detections:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  cv2.rectangle(image, (l,t), (r, b), (255, 255, 0), 2 )

cv2_imshow(image)
```
